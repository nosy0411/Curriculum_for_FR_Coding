```python
%matplotlib inline

import FinanceDataReader as fdr
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

price_data = fdr.DataReader('005930', '2020-01-01')
KOSPI_data = fdr.DataReader('KS11', '2020-01-01')

#start = '2020-01'
#end = '2020-09'

#draw fig and axis 1
fig, ax = plt.subplots(figsize=(10, 6), facecolor = 'white')
ax.plot(KOSPI_data['Close'],lw = 3, color = 'r',label='KOSPI')
ax.axis('tight')                          
ax.legend(loc=0)                             
#date format
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
#axis 2
ax2 = ax.twinx()  
ax2.plot(price_data['Close'],lw=1, color='b', label='Samsung Electronics')
ax2.legend(loc=0)  
#grid option
plt.grid(True, axis='y')
#set label
ax.set_ylabel('KOSPI')  
ax2.set_ylabel('Samsung Electronics')            

#       .(,  )
#price_data

```
