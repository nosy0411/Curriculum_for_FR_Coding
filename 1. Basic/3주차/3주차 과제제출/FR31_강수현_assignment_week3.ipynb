```python
%matplotlib inline

import matplotlib.dates as mdates
from datetime import datetime


price_data = fdr.DataReader('005930', '2020-01-01')
KOSPI_data = fdr.DataReader('KS11', '2020-01-01')

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(KOSPI_data['Close'])
plt.plot(KOSPI_data['Close'], color ='r', lw=3, label ='KOSPI')
plt.ylabel('KOSPI')
plt.legend(loc=0)
plt.grid(True)
plt.axis('tight')
#    
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1)) 

fig, 
ax2 = ax.twinx() 
ax2.plot(price_data['Close'])
plt.plot(price_data['Close'], color ='b', lw=1, label ='Samsung Electronics')
plt.ylabel('Samsung Electronics')
plt.legend(loc=0)
```
