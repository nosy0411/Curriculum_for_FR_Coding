```python
%matplotlib inline
# 주가를 손쉽게 불러올 수 있는 라이브러리를 호출합니다.
import FinanceDataReader as fdr
import numpy as np
import matplotlib.pyplot as plt
# 2020년부터 오늘까지의 005930(삼성전자) 주가 데이터를 모두 불러옵니다.
# 4주차에 배울 데이터프레임 형식으로 불러오기 때문에 관련 코드는 아래 코드 그대로 사용하시면 됩니다.
# 만약 예습 겸 데이터프레임 다루는 코드를 알고 싶다면, 4주차 예습을 권장합니다. (필수 X)
price_data = fdr.DataReader('005930', '2020-01-01')
KOSPI_data = fdr.DataReader('KS11', '2020-01-01')

# 데이터프레임을 확인해 볼 수 있는 테스트 코드입니다.(참고용, 지워도 무방)
price_data

# size는 10, 6 입니다.
# 데이터프레임의 특정 열은 아래와 같은 방식으로 접근합니다.
# ?.plot(KOSPI_data['Close'])
# ?.plot(price_data['Close'])

# KOSPI의 선 두께는 3으로 하고, 범례는 최적위치로 합니다.
```

```python
fig, ax = plt.subplots(figsize=(10, 6)) 
ax.plot(KOSPI_data['Close'], color='r', lw=3, label='KOSPI')
ax.set_ylabel('KOSPI')

ax2 = ax.twinx()                              
ax2.plot(price_data['Close'], color='b', label='Samsung Electronics') 
ax2.yaxis.grid(True)
ax2.set_ylabel('Samsung Electronics')                   
fig.legend(loc=0)
```
