```python
import FinanceDataReader as fdr
import numpy as np
import matplotlib.pyplot as plt

price_data = fdr.DataReader('005930', '2020-01-01')
KOSPI_data = fdr.DataReader('KS11', '2020-01-01')
fig = plt.figure(figsize=(10, 6))              
ax = fig.subplots()                            

a=ax.plot(KOSPI_data['Close'], lw=3, color='r',label='KOSPI')
ax.grid(axis='y')
ax.axis('tight')                               
                           
ax.set_xlabel('')                           
ax.set_ylabel('KOSPI')                  


ax2 = ax.twinx()                                
b=ax2.plot(price_data['Close'], lw=1, color='b',label='Samsung Electronics') 
ax2.set_ylabel('Samsung Electronics')                      


fig.legend()
fig.show()

```
