```python
import FinanceDataReader as fdr
import numpy as np
import matplotlib.pyplot as plt
 

KOSPI_data = fdr.DataReader('KS11', '2020-01-01')
price_data = fdr.DataReader('005930', '2020-01-01')


fig = plt.figure(figsize=(10, 6))
ax = fig.subplots() 


ko = ax.plot(KOSPI_data['Close'], label='KOSPI', color='r', lw=3)
ax2 = ax.twinx()
sa = ax2.plot(price_data['Close'], label='Samsung Electronics', color='b', lw=1)

re = ko + sa
labs = [l.get_label() for l in re]
ax.legend(re, labs, loc=0)


ax.grid(True, axis='y')
ax.set_ylabel('KOSPI')
ax2.set_ylabel('Samsung Electronics')                    


plt.show()



```
