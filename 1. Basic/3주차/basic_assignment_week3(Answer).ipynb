```python
# %matplotlib inline
# 주가를 손쉽게 불러올 수 있는 라이브러리를 호출합니다.
import FinanceDataReader as fdr
import numpy as np
import matplotlib.pyplot as plt
# 2020년부터 오늘까지의 005930(삼성전자) 주가 데이터를 모두 불러옵니다.
# 4주차에 배울 데이터 프레임 형식으로 불러오기 때문에 관련 코드는 아래 코드 그대로 사용하시면 됩니다.
price_data = fdr.DataReader('005930', '2020-01-01')
KOSPI_data = fdr.DataReader('KS11', '2020-01-01')
print(price_data)


fig = plt.figure(figsize=(10, 6))    
ax = fig.subplots()
ax.plot(KOSPI_data['Close'], lw=3, color='r', label='KOSPI')             
ax.set_ylabel('KOSPI')

ax2 = ax.twinx()
ax2.plot(price_data['Close'], color='b', label='Samsung Electronics')
ax2.set_ylabel('Samsung Electronics')
ax2.grid(True) 

fig.legend()   
plt.show()
```
